# Step 4 Loyalty Flag

```python
# Use groupby and transform() functions with np.max argument to determine maximum order number for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Use loc function to make loyalty flag column
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking the frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#Check the head to make sure our coumn has been added succesfully (it has)
ords_prods_merge.head()
```

# Step 5 Spending Habbits

# Step 6 Spending flag for each user

### Low Spender / High Spender

```python
# apply the "transform function here to determine the mean of "user_id" and "prices"
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check the head to see it was succesfull and then to observe and make comment on the data
ords_prods_merge[['user_id', 'mean_product_price']].head(15)
```

```python
#Creating the flag for the low spenders
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#creating a flag for the high spenders
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Check the frequency of the new column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#Checking the head with the above Crteria
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head(15)
```

# Step 7 Frequent vs Non-Frequent users

```python
# us the group by and transform togehter to ge the median
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check the head to make sure it was succersfull
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(15)
```

```python
#create the frequency flags using the loc. function and the given criteria
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
#Check the frequency
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```
