### Creating the "busiest_day" column

```python
# here we use the results to define what the days are
#the results go from 0 - 6 this is not the ranking but the weekday.
#because python views 0 as a value the seven weekdays will be betweem 0 and 6.
#0 monday thought to 6 sunday
#the order the results are displayed in are the rank made evident by the frequency displayed next to the value.
result = []

for value in ords_prods_merge["order_day"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

### I will define the rest as "Regularly busy"

```python
#i will define these rulsts as results_2 so as to not get mixed up and to not change the results of step 1
result_2 = []

for value in ords_prods_merge["order_day"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

# Step 4 Creating a busiest_period_of_day. column

```python
# checking thr rank of the order hour of the day
ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
#because we are creatring 3 categories namely Most orders, Average orders, and Fewest orders.
#we can divide the 24 hours of the day by 3. top 6, middle 6 and bottom 6 respectively.
```

```python
result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
#creating the new column "busiest_period_of_day"
ords_prods_merge['busiest_period_of_day'] = result_3
```
